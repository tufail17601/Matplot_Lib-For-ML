```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']
language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))

languages = []
popularity = []

for item in language_counter.most_common(10):
    languages.append(item[0])
    popularity.append(item[1])
languages.reverse()
popularity.reverse()
plt.style.use('fivethirtyeight')
colors = ['coral'] + ['skyblue'] * (len(languages) - 1)

plt.barh(languages,popularity,color = ['skyblue',
    'orange',
    'limegreen',
    'crimson',
    'gold',
    'purple',
    'teal',
    'coral',
     'slateblue',
    'tomato'])
plt.xlabel('popularity')
plt.ylabel('Languages')
plt.title('Languages Used in the survay')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('salaries.csv')
age = df['Age']
dev_salaries = df['All_Devs']
py_dev_salaries = df['Python']
java_dev_salaries = df['JavaScript']
plt.style.use('fivethirtyeight')
plt.plot(age,dev_salaries,label = 'All_devs',color= 'red',linestyle = '--')
plt.plot(age,py_dev_salaries,label= 'Python',color = 'green')
median = 57287
plt.fill_between(age, py_dev_salaries, dev_salaries,
                 where=(py_dev_salaries > dev_salaries),
                 interpolate=True, alpha=0.5, label='Python > All Devs')

# Fill between where Python <= All Devs
plt.fill_between(age, py_dev_salaries, dev_salaries,
                 where=(py_dev_salaries <= dev_salaries),
                 interpolate=True, alpha=0.5, color='red', label='Python < All Devs')
plt.title('Salaries of Developers by Age')
plt.xlabel('AGES')
plt.ylabel('Salaries')
plt.legend()
plt.tight_layout()
plt.show()
#df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('2019-05-31-data (1).csv')

# Extract relevant columns
view = data['view_count']
likes = data['likes']
ratio = data['ratio']

# Create scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(view, likes, alpha=0.88,
                      c=ratio, cmap='summer', edgecolors='black', s=120)

# Color bar to represent ratio
cbar = plt.colorbar(scatter)
cbar.set_label('Like/Dislike Ratio')

# Log scale for better distribution
plt.xscale('log')
plt.yscale('log')

# Add labels and title
plt.title('YouTube Video Likes vs Views (Log Scale)', fontsize=14)
plt.xlabel('View Count (log scale)')
plt.ylabel('Likes (log scale)')

# Add grid
plt.grid(True, which="both", linestyle='--', linewidth=0.5)

# Make layout clean
plt.tight_layout()
plt.show()
```
